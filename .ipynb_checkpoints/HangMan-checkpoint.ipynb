```python
import random

def hangman():

    with open('/usr/share/dict/words') as file:
        words = file.readlines()
        word = random.choice(words).strip().lower()
        print('the word is ' + word)

    guessed_letters = []
    attempts = 6
    
    print('Let\'s play HangMan! \n')
    
    guessed_word = '_' * len(word)

    while attempts > 0:
        print(' '.join(list(guessed_word)))
        guess = input('Enter a letter: ').lower()
        guessed_letters.append(guess)
        
        
        if guess in guessed_letters:
            print('You already guessed that letter! Try again!')
        
        else if guess in word:
            for index, char in enumerate(word):
                if char == guess:
                    guessed_word = guessed_word[:index] + guess + guessed_word[index + 1:]
#                     this loop runs through the word and when it hits a letter 
#                     that is in the word, it stops and replaces the letter in 
#                     the word with your guessed char, then moves up one index 
#                     and continues on until it either ends or doesnt find another
#                     index of the guessed letter
                    
                    
            if '_' not in guessed_word:
                print('Congratulations! You guessed the word!')
                return
        else:
            attempts = attempts - 1
            if attempts == 0:
                print('You lose! You ran out of guesses!')
                
        print('Guessed Letters: ' + ' '.join(guessed_letters))
            
        
            
    
hangman()    
```

```python
import random

def hangman():
    
    with open('/usr/share/dict/words') as file:
        words = file.readlines()
        word = random.choice(words).strip().lower()
        
    incorrect_guesses = 0
    
    image_list = 
              '    +---+'
              '    |   |'
              '    O   |'
              '   /|\  |'
              '   / \  |'
              '        |'
              '  ========='
    
    word_list = list(word) 
    board = ["__"] * len(word)
    
    win = False
    
    print('welcome to hangman')
    
    while incorrect_guesses < len(image_list) - 1:
        print('\n')
        msg = 'Guess a letter:'
        guess = input(msg)
        
        if guess in word_list:
            for index, char in enumerate(word_list):
                if char ==guess:
                    cind = word_list.index(guess)
                    board[cind] = char
                    word_list[cind] = '$'
        else:
            incorrect_guesses += 1
        
        print("".join(board))
        
        end = incorrect_guesses + 1
        print("\n".join(image_list[0: end]))
        
        if "__" not in the board:
            print('you win')
            print(" ".join(board))
            win = True
            break
            
    if not win:
        print(f'u lose, the word was {word}.')
    
hangman()
```
